This Project is to make a linear reression model that can determine or predict the compressive strength of concrete

```python
# Importing libraries
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd
import joblib
# Assigning the dataset to a variable and understanding the data

import pandas as pd
concrete_data = pd.read_excel("Concrete_Data.xls")
concrete_data.head()

concrete_data.tail()

concrete_data.shape
concrete_data.columns

concrete_data.info()

concrete_data.describe()
concrete_data.isnull().sum()

# Assigning each column to a variable

x = concrete_data.drop('Concrete compressive strength(MPa, megapascals) ' , axis=1)
y = concrete_data['Concrete compressive strength(MPa, megapascals) ']

x.shape

# spliting the data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x,y,
    test_size=0.2,
    random_state=42
)

# Training the model
model = LinearRegression()
model.fit(x_train, y_train)
joblib.dump(model, "Concrete_model.pkl")

y_pred = model.predict(x_test)


# Model Evaluation

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:" , mse)
print("R^2:" , r2)

# Plotting actual v predicted
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)

import joblib

joblib.dump(model, "Concrete_model.pkl")






```
